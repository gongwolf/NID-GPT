```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
csv1_path = '/data/qgong/data/CICDDOS2019/results/ddpm_synthetic_data_all.csv'
csv2_path = '/data/qgong/data/CICDDOS2019/training.csv'

# csv1_path = '/data/qgong/data/UNSW/results/ddpm_synthetic_data_all.csv'
# csv2_path = '/data/qgong/data/UNSW/training_all_classes.csv'

data1 = pd.read_csv(csv1_path)
data2 = pd.read_csv(csv2_path)
```

```python
import random

num_columns = data1.shape[1]

# Select 4 random column indices
random_columns = random.sample(range(num_columns), 4)

# Create a figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid for 4 plots
axes = axes.flatten()  # Flatten the axes for easier iteration

# Iterate over the selected columns and draw on subplots
for i, column_index in enumerate(random_columns):
    ax = axes[i]
    ax.hist(data1.iloc[:, column_index], 
            bins=30, 
            alpha=0.8,
            label='ddpm',
            density=True,
            color='#1f77b4',  # Dark blue color
            edgecolor='black')
    ax.hist(data2.iloc[:, column_index], 
            bins=30, 
            alpha=0.5,
            label='train',
            density=True,
            color='#ff7f0e',  # Light orange/pink color
            edgecolor='black')
    
    # Add labels, legend, and title
    ax.set_xlabel('Value')
    ax.set_ylabel('Density')
    ax.set_title(f'Distribution of Column {column_index}')
    ax.legend()
    ax.grid()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
